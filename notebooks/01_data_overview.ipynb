```python
transfers = pd.read_csv('../data/transfers.csv')
transfers['transfer_date'] = pd.to_datetime(transfers['transfer_date'])
cutoff = pd.Timestamp('2026-07-01')
before = len(transfers)
transfers = transfers[transfers['transfer_date'] <= cutoff]
after = len(transfers)
print(f"Removed {before-after} transfers (after 2026-07-01)")
transfers.to_csv('../data/transfers.csv', index=False)
```
